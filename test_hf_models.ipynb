```python
import json
with open("EN2001a.json", "r") as f:
    data = json.load(f)
start = data['output']['diarization'][0]['start']
end = data['output']['diarization'][0]['end']

audio_path = "amicorpus/EN2001a/audio/EN2001a.Mix-Headset.wav"
```

Testing embedding model and similarity scores.

```python
import os
from pyannote.audio import Model, Inference

model = Model.from_pretrained("pyannote/embedding", token=os.getenv("HF_TOKEN"))
inference = Inference(model)

# inference on an excerpt
from pyannote.core import Segment
excerpt = Segment(start=start, end=end)
inference.crop(audio_path, excerpt)
```
